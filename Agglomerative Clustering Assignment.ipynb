```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#  Load iris data from load_iris function from sklearn.datasets package.
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
# From the dataset extract the data property.
iris_data = pd.DataFrame(iris['data'], columns = iris['feature_names'] )
iris_data.head()
```

```python
import seaborn as sns

sns.pairplot(iris_data)
plt.show()
```

```python
# Train an AgglomerativeClustring model based on the data.
iris_d = iris_data[['petal length (cm)', 'petal width (cm)']]

from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
ac.fit_predict(iris_d)
```

```python
plt.scatter(iris_data['petal length (cm)'] ,iris_data['petal width (cm)'], c= ac.labels_, cmap='cool')
```

```python
# Plot dendrogram to visualize the clustering linkage

from scipy.cluster.hierarchy import dendrogram, linkage
dn = dendrogram(linkage(iris_d, method = 'ward'))
```
